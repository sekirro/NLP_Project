```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('orpo.csv')

x = df['Step']
y = df['Training Loss']

plt.plot(x, y)

# 添加标签和标题
plt.xlabel('Step')
plt.ylabel('Training Loss')
plt.title('Training Loss')
plt.legend()  # 显示图例

# 添加网格线（可选）
plt.grid(True, linestyle='--', alpha=0.7)

plt.show()
```

```python
# 读取数据
df = pd.read_csv('orpo.csv')

# 创建图表
plt.figure(figsize=(10, 6))  # 设置图表大小

# 绘制两条线
plt.plot(df['Step'], df['rewards / chosen'], label='Chosen', color='blue')
plt.plot(df['Step'], df['rewards / rejected'], label='Rejected', color='red')

# 添加标签和标题
plt.xlabel('Step')
plt.ylabel('Rewards')
plt.title('Chosen vs Rejected Rewards')
plt.legend()  # 显示图例

# 添加网格线（可选）
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图表
plt.show()
```

```python
# 读取数据
df = pd.read_csv('orpo.csv')

# 计算差值
df['rewards_margin'] = df['rewards / chosen'] - df['rewards / rejected']

# 创建图表
plt.figure(figsize=(10, 6))  # 设置图表大小

# 绘制差值线
plt.plot(df['Step'], df['rewards_margin'], label='Chosen - Rejected', color='purple')

# 添加零线参考
plt.axhline(y=0, color='gray', linestyle='--', alpha=0.5)

# 添加标签和标题
plt.xlabel('Step')
plt.ylabel('Rewards Margin (Chosen - Rejected)')
plt.title('Rewards Margin (Chosen - Rejected)')
plt.legend()  # 显示图例

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 显示图表
plt.show()
```
